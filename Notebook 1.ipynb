```
# Start writing code here...
import random 
import math

lower  = int(input('Enter a lower bound: '))
upper  = int(input('Enter a upper bound: '))
x = random.randint(lower, upper)

count = 0
while count < math.log(upper - lower + 1, 2):
    count += 1

guess = int(input('Guess a number'))
if guess == x:
    print('Congratulations you have guessed right', "try")
elif x > guess:
    print ('Guess is too small, try again')
elif x < guess:
    print ('Guess is too big, try again')


if count >= math.log(upper - lower + 1, 2):
    print("\nThe number is %d" % x)
    print("\tBetter Luck Next time!")
```

```
import random
options = ['Rock', 'Paper', 'Scissors']
name = input("Enter your name:")
Playerscores = 0
Computerscores = 0
Noofrounds = 0

Gameon = True
print(f"Welcome {name.title()}")
while Noofrounds<5:
    ComputerOptions = random.choice(options)
    PlayerOptions = input("Enter Rock/Paper/Scissors: ").title()
    print(f"Computer's Option: {ComputerOptions}")
    print(f"Player's Option: {PlayerOptions}")

    Noofrounds += 1
    if ComputerOptions==PlayerOptions:
         print('Tie')
    elif (ComputerOptions == 'Rock' and PlayerOptions == 'Scissors') or (ComputerOptions == 'Scissors' and PlayerOptions == 'Paper') or (ComputerOptions == 'Paper' and PlayerOptions == 'Rock'):
        print('Computer Wins')
        Computerscores += 1
    elif (PlayerOptions == 'Rock' and ComputerOptions == 'Scissors') or (PlayerOptions == 'Scissors' and ComputerOptions == 'Paper') or (PlayerOptions == 'Paper' and ComputerOptions == 'Rock'):
        print('Player Wins')
        print(f"{name.title()} wins")
        Playerscores+=1
    else:
        print("Enter a valid option")
    print("-------------------------")
    print("")
    print(f"Round No: {Noofrounds}")
    print("------ Score Board ------")
    print(f"{name.title()}: {Playerscores} | Computer: {Computerscores}")
    print("===============================")
    print("")
    if Noofrounds==5:
        gameOn=False
        break
if Playerscores==Computerscores:
  print("Draw!!")
elif Playerscores>Computerscores:
  print(f"Congrats {name.title()}, You won the game!!")
else:
  print(f"Oops Computer won the game!! Better luck next time {name.title()}!")
```

```
#TIMER 

import time

def countdown(t):
    while t:
        mins, secs  = divmod(t, 60)
        timer = '{:02d}:{:02d}'.format(mins, secs)
        print(timer, end="/r")
        time.sleep(1)
        t-=1
    print('Timer Completed')

t = input("Length of Timer in Seconds")
countdown(int(t))



```
